**MatplotLib**

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy.

Major pros of Matplotlib are:
    . Generally easy to get started for simple plots
    . Support for custom labesl and texts
    . Great control of every element in a figure
    . High-quality output in many formats
    . Very customizable in general

```python
## plt plot  
# plt.plot(x,y,'g--',linestyle='dashed',linewidth=2,markersize=2)
plt.plot(x,y,'g--')  
plt.xlabel('X axis')
plt.ylabel('Y axis')   
plt.title('2D Graph')
```

```python
## Creating  Subplots
plt.subplot(2,2,1)
plt.plot(x,y,'g--')
plt.subplot(2,2,2)
plt.plot(x,y,'r*-')
plt.subplot(2,2,3)
plt.plot(x,y,'o')
plt.subplot(2,2,4)
plt.plot(x,y,'y--')
```

```python
# Compute the x and y coordinates for points on a sine curve
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("sine wave form")

# plot the points using matplotlib
plt.plot(x,y)
# plt.show()
```

```python
#Subplot
# Compute the x and y coordinates for points on a sine and cosine curves
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

# Set up a subplot grid that has height 2 and width 1,
# and set the first such subplot as active
plt.subplot(2,1,1)

# Make the first plot
plt.plot(x,y_sin)
plt.title('Sine')

# Set the seond subplot as active, and make the seond plot
plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('Cosine')

```

```python
## Bar Plot
x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y)
plt.bar(x2,y2,color='g')        ## color is mandatory 
plt.title('Bar graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

**Histograms**

```python
a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title("histogram")
```

**Box Plot using Matplotlib**

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]

#rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);  ## patch_artist represents the color if False
```

**Pie Chart**

```python
# Data to plot
labels='Python','C++','Ruby','Java'
sizes=[215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0,0,0) # explode 1st slice

# Plot
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
```
